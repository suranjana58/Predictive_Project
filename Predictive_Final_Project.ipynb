```python
######### MISSING VALUE IMPUTATION #############
# Ideally, there should be 3652 days. So we need to impute missing values for rest of the cities
list_Cities = list()
final_df = pd.DataFrame()
```

```python
list_Cities = ["Bakersfield, CA",
"Bangor, ME",
"Cheyenne, WY",
"Fresno, CA",
"Kahului-Wailuku-Lahaina, HI",
"Los Angeles-Long Beach-Anaheim, CA",
"San Francisco-Oakland-Hayward, CA",
"Urban Honolulu, HI",
"Visalia-Porterville, CA",
"Wilmington, NC"]
```

```python
start_date = '2013-01-01'
end_date = '2022-12-31'
date_range = pd.date_range(start=start_date, end=end_date)
all_dates_df = pd.DataFrame({'Date_2': date_range})


for i in list_Cities:
    Temp = Summary[Summary["CBSA_NAME"]==i]
    
    # Merge the two DataFrames to add missing days
    df = Temp.append(all_dates_df)
    df["CBSA_NAME"] = i
    df = df.drop_duplicates(subset=['Date_2'], keep='first')
    del Temp
    final_df = final_df.append(df)
```

```python
Check_2 = final_df.groupby(['CBSA_NAME']).agg(Count=('CBSA_NAME', 'count')).reset_index()
Check_2.head(10)
```

```python
final_df = final_df.sort_values(['CBSA_NAME', 'Date_2'])
final_df['PM25_Level_Imputed'] = final_df['PM25_Level'].rolling(window=5, min_periods=1).apply(lambda x: np.mean(x))
```

```python
# Might not have to run it everytime
final_df.drop(['level_0', 'index'], axis=1, inplace=True)
```

```python
final_df['PM25_Level_FINAL'] = final_df['PM25_Level'].fillna(final_df['PM25_Level_Imputed'])
final_df
```
